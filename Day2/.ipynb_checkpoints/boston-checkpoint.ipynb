### 1. データを準備する、確認する

```python
#必要なライブラリを読み込む
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#ボストン住宅価格のデータセットを読み込む
from sklearn.datasets import load_boston
boston_data = load_boston()
```

```python
#データセットの中身を確認する
print("==================DESCR====================")
print(boston_data.DESCR)
print("==================DATA=====================")
print(boston_data.data)
print("===============FEATURE_NAME================")
print(boston_data.feature_names)
print("=================TARGET====================")
print(boston_data.target)
```

```python
#説明変数となるデータを準備する
df_X = pd.DataFrame(data=boston_data.data, columns=boston_data.feature_names)
df_X
```

```python
#目的変数となるデータを準備する
df_y = pd.DataFrame(data=boston_data.target, columns=["MEDV"])
df_y
```

```python
#基本統計量
df_X.describe()
```

```python
#変数の数、データ数、データ型、NULL値の有無を確認する
df_X.info()
```

```python
#データをまとめて、相関関係を見る
df_all = pd.concat([df_X, df_y], axis=1)
df_all.corr()
```

```python
#相関関係を可視化する
plt.figure(figsize=(15,10))
sns.heatmap(df_all.corr(),annot=True)
```

```python
#相関の強い要素のデータの関係性を可視化する
sns.jointplot(x="RM",y="MEDV",data=df_all.loc[:,["RM","MEDV"]])
```

### 2. 単回帰モデルを作る

```python
#学習に必要なデータを抽出する
df_RM = df_all.loc[:,["RM","MEDV"]]
```

```python
#説明変数を作る
X = df_RM.RM.values.reshape((len(df_RM.RM),1))
X
```

```python
#目的変数を作る
y = df_RM.MEDV
y
```

```python
#学習データとテストデータを作る
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
print(len(X_train))
print(len(X_test))
```

```python
#単回帰モデルを作る
from sklearn.linear_model import LinearRegression as LR
reg = LR()
#学習する
model = reg.fit(X_train, y_train)
```

```python
#予測結果を可視化する
plt.scatter(X_test, y_test)
plt.plot(X_test, model.predict(X_test))
plt.show()
```

```python
model.predict(10)
```

### 3. 重回帰モデルを作る

```python
#学習データとテストデータを作る
X = df_X.values
y = df_y.values

X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(X, y, test_size=0.33)
```

```python
#学習する
reg_multi = LR()
model_multi = reg_multi.fit(X_train_multi, y_train_multi)
```

```python
#単回帰モデルと重回帰モデルを評価する（エラー率）
from sklearn.metrics import mean_squared_error

print("liner:%.3f" %mean_squared_error(y_test,model.predict(X_test)))
print("multi:{}" .format(mean_squared_error(y_test_multi,model_multi.predict(X_test_multi))))
```
